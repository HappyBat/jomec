```python
# import the .csv and create a pd df

path_to_file = "/Users/aidanair/Documents/DATA/ALL_DATASETS/canada_fb_ads_main.csv"
df = pd.read_csv(path_to_file, sep=',', escapechar = '\\', encoding = 'utf-8')

# this didn't work error-free until we put in the 'escapechar' parameter.

# if it comes to it we can:
# data = pd.read_csv(path_to_file, error_bad_lines=False)
```

```python
df[16:19]
```

```python

# flip two columns to datetime via .apply (which maps a function across a given index - here the 2 columns
# which are currently strings)

df[['ad_creation_time', 'ad_delivery_start_time']] = df[['ad_creation_time', 'ad_delivery_start_time']].apply(pd.to_datetime)
```

```python
# check

df.info()
```

```python
# now deal with the stoptime column, which is also a string format>
# Why is it done via an 'if' clause?


if 'ad_delivery_stop_time' in df.columns:
    df.ad_delivery_stop_time = pd.to_datetime(df.ad_delivery_stop_time)
```

```python
# "Calculate median of impression and ad spend ranges to get a more realistic estimate"
# i.e. build a new column (spend_median) which takes the two spend columns and maps 'np.median' across them

df['spend_median'] = df[['spend_lower_bound', 'spend_upper_bound']].apply(np.median, axis = 1)
```

```python
df['spend_median']
```

```python
# as previously but with impressions rather than spend

df['impressions_median'] = df[['impressions_lower_bound', 'impressions_upper_bound']].apply(np.median, axis = 1)
```
