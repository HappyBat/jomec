## VLOOKUP

```python
# show me a count of the different elements in a given column

listings_summ.room_type.value_counts()
```

```python
# create a dictionary of keys (existing) and values (what you want instead)

room_rename_dict = {"Entire home/apt":"the lot", "Private room":"just the room", "Hotel room": "Hotel", "Shared room": "sharing"}
```

```python
# use .map to map the values into a new column with the following syntax:
# df['new column name'] = df.existing_column.map(dictionary of keys and values)

listings_summ["kind_of_room"] = listings_summ.room_type.map(room_rename_dict)
```

### Filtering & sorting

```python
# use matplotlib to make a pie chart AND add a circle

import matplotlib.pyplot as plt
 
# Create a pieplot
plt.pie(listings_summ.room_type.value_counts())
 
# add a circle at the center - 2nd argument is the white circle radius
my_circle = plt.Circle((0,0), 0.7, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)
 
plt.show()
```
